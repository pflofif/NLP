```python
import json
from src.preprocess import preprocess

def run_tests():
    passed = 0
    with open('../tests/edge_cases.jsonl', 'r', encoding='utf-8') as f:
        for line in f:
            case = json.loads(line)
            result = preprocess(case['input'])['clean_full']
            
            if result == case['expected']:
                passed += 1
            else:
                print(f"FAIL: In: {case['input']} | Expected: {case['expected']} | Got: {result}")
    
    print(f"Result: {passed} tests passed.")

run_tests()
```
